```python
# =============================================================================
# COMPENDIO DE EJERCICIOS: INDEXACIÓN, FILTRADO Y AGRUPACIÓN EN PANDAS
# =============================================================================

import pandas as pd
import numpy as np

# Semilla para reproducibilidad
np.random.seed(42)
```

```python
# -----------------------------------------------------------------------------
# EJERCICIO 1: Selección básica con .loc (etiquetas)
# -----------------------------------------------------------------------------
# Dataset de ventas de productos
fechas = pd.date_range('2024-01-01', periods=10, freq='D')
productos = ['A', 'B', 'C', 'D', 'E']
df_ventas = pd.DataFrame({
    'fecha': np.random.choice(fechas, 20),
    'producto': np.random.choice(productos, 20),
    'cantidad': np.random.randint(1, 10, 20),
    'precio': np.round(np.random.uniform(10, 100, 20), 2)
})
df_ventas.set_index('fecha', inplace=True)
df_ventas.sort_index(inplace=True)

print("=== DATASET VENTAS (indexado por fecha) ===")
print(df_ventas.head(10))
```

```python
# -----------------------------------------------------------------------------
# Tareas con .loc
# -----------------------------------------------------------------------------
# Tarea 1.1: Selecciona todas las filas desde '2024-01-03' hasta '2024-01-05' inclusive.
# Resultado esperado: DataFrame con fechas en ese rango.

# Tarea 1.2: Selecciona las filas con fecha '2024-01-04' y solo las columnas 'producto' y 'cantidad'.
# Resultado esperado: DataFrame con una fila y dos columnas.

# Tarea 1.3: Selecciona todas las filas donde el producto sea 'A' o 'C' y la cantidad sea mayor a 5.
# Resultado esperado: Filas que cumplan ambas condiciones.

# Tarea 1.4: Usando .loc, asigna el valor 99.99 al precio de todas las filas del producto 'B'.
# Resultado esperado: El precio del producto 'B' cambia a 99.99.
```

```python
# Tarea 1.1: Selecciona todas las filas desde '2024-01-03' hasta '2024-01-05' inclusive.
# Resultado esperado: DataFrame con fechas en ese rango.
df_ventas.loc[(df_ventas.index>="2024-01-03") & (df_ventas.index<="2024-01-05")]
```

```python
df_ventas[df_ventas.index=="2024-01-04"][["producto","cantidad"]]
```

```python
# -----------------------------------------------------------------------------
# EJERCICIO 3: Agrupación básica con .groupby y agregación simple
# -----------------------------------------------------------------------------
# Dataset de ventas por categoría y región
df_ventas2 = pd.DataFrame({
    'region': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], 30),
    'categoria': np.random.choice(['Electrónica', 'Ropa', 'Hogar'], 30),
    'ventas': np.random.randint(100, 1000, 30),
    'utilidad': np.random.randint(10, 200, 30)
})
print("\n=== DATASET VENTAS POR REGIÓN Y CATEGORÍA ===")
print(df_ventas2.head(10))

# -----------------------------------------------------------------------------
# Tareas con groupby
# -----------------------------------------------------------------------------
# Tarea 3.1: Agrupa por 'region' y calcula la suma total de 'ventas' por región.
# Resultado esperado: Serie o DataFrame con una fila por región y la suma de ventas.

# Tarea 3.2: Agrupa por 'categoria' y calcula el promedio de 'utilidad' por categoría.
# Resultado esperado: Promedio de utilidad para Electrónica, Ropa y Hogar.

# Tarea 3.3: Agrupa por 'region' y 'categoria' y obtén el número de transacciones (count) por grupo.
# Resultado esperado: Tabla con regiones y categorías como índices múltiples y el conteo.
```

```python
# Tarea 3.1: Agrupa por 'region' y calcula la suma total de 'ventas' por región.
# Resultado esperado: Serie o DataFrame con una fila por región y la suma de ventas.
```

```python
print(df_ventas2.groupby(by = "region").sum())
```

```python
# Tarea 3.2: Agrupa por 'categoria' y calcula el promedio de 'utilidad' por categoría.
# Resultado esperado: Promedio de utilidad para Electrónica, Ropa y Hogar.
```

```python
print(df_ventas2.groupby(by = ["categoria"])["utilidad"].mean())
```

```python
# Tarea 3.3: Agrupa por 'region' y 'categoria' y obtén el número de transacciones (count) por grupo.
# Resultado esperado: Tabla con regiones y categorías como índices múltiples y el conteo.
```

```python
print(df_ventas2.groupby(by = ['region','categoria']).count())
```

```python
# -----------------------------------------------------------------------------
# EJERCICIO 4: Agregaciones múltiples con .agg
# -----------------------------------------------------------------------------
# Mismo dataset df_ventas2

# Tarea 4.1: Agrupa por 'region' y aplica .agg para obtener: suma de ventas, promedio de utilidad,
# y desviación estándar de ventas. Renombra las columnas a 'total_ventas', 'utilidad_promedio', 'std_ventas'.
# Resultado esperado: DataFrame con regiones como índice y las tres métricas.

# Tarea 4.2: Agrupa por 'categoria' y aplica funciones personalizadas con .agg:
# - Rango de ventas (max - min)
# - Mediana de utilidad
# Resultado esperado: DataFrame con categorías y las dos nuevas métricas.

# Tarea 4.3: Usa named aggregation para agrupar por 'region' y calcular:
#   ventas_max = max de ventas
#   utilidad_min = min de utilidad
#   ventas_mean = media de ventas
# Resultado esperado: DataFrame con regiones y columnas con esos nombres.
```

```python
# Tarea 4.1: Agrupa por 'region' y aplica .agg para obtener: suma de ventas, promedio de utilidad,
# y desviación estándar de ventas. Renombra las columnas a 'total_ventas', 'utilidad_promedio', 'std_ventas'.
# Resultado esperado: DataFrame con regiones como índice y las tres métricas.
```

```python
resultado = df_ventas2.groupby('region').agg(
    total_ventas=('ventas', 'sum'),
    std_ventas=('ventas', 'std'),
    utilidad_promedio=('utilidad', 'mean')
)

print(resultado)
```

```python
# Tarea 4.2: Agrupa por 'categoria' y aplica funciones personalizadas con .agg:
# - Rango de ventas (max - min)
# - Mediana de utilidad
# Resultado esperado: DataFrame con categorías y las dos nuevas métricas.
```

```python
print( df_ventas2.groupby('categoria').agg(
        rango_ventasMIN = ('ventas','min'),
        rango_ventasMAX = ('ventas','max'),
        mean_utility = ('utilidad','mean')
))
```

```python
# Tarea 4.3: Usa named aggregation para agrupar por 'region' y calcular:
#   ventas_max = max de ventas
#   utilidad_min = min de utilidad
#   ventas_mean = media de ventas
# Resultado esperado: DataFrame con regiones y columnas con esos nombres.
```

```python
print( df_ventas2.groupby('region').agg(
        rango_ventasMAX = ('ventas','max'),
        utilidadMIN = ('utilidad','min'),
        sales_utility = ('ventas','mean')
))
```

```python
# -----------------------------------------------------------------------------
# EJERCICIO 5: Filtrado de grupos con .filter
# -----------------------------------------------------------------------------
# Dataset de empleados
df_empleados = pd.DataFrame({
    'departamento': np.random.choice(['RH', 'IT', 'Ventas', 'Marketing'], 20),
    'salario': np.random.randint(30000, 90000, 20),
    'edad': np.random.randint(22, 60, 20),
    'experiencia': np.random.randint(0, 30, 20)
})
print("\n=== DATASET EMPLEADOS ===")
print(df_empleados)

# -----------------------------------------------------------------------------
# Tareas con .filter
# -----------------------------------------------------------------------------
# Tarea 5.1: Filtra los departamentos cuyo salario promedio sea mayor a 55000.
# Resultado esperado: DataFrame completo solo con filas de los departamentos que cumplen.

# Tarea 5.2: Filtra los departamentos que tengan al menos 5 empleados.
# Resultado esperado: DataFrame con departamentos que cumplen la condición.

# Tarea 5.3: Filtra los departamentos donde la edad máxima sea menor a 50.
# Resultado esperado: DataFrame con departamentos cuya edad máxima < 50.
```

```python
# Tarea 5.1: Filtra los departamentos cuyo salario promedio sea mayor a 55000.
# Resultado esperado: DataFrame completo solo con filas de los departamentos que cumplen.
```

```python
df_empleados_top = (
    df_empleados
    .groupby('departamento')
    .filter(lambda x: x['salario'].mean() > 55000) # <--- El .filter va ANTES, sin el .agg()
)

print(df_empleados_top)
```

```python
# Tarea 5.2: Filtra los departamentos que tengan al menos 5 empleados.
# Resultado esperado: DataFrame con departamentos que cumplen la condición.
```

```python
df_empleados_top = df_empleados.groupby('departamento').count()


print(df_empleados_top.loc[(df_empleados_top['edad']>5)].index.values)
```

```python
# Tarea 5.3: Filtra los departamentos donde la edad máxima sea menor a 50.
# Resultado esperado: DataFrame con departamentos cuya edad máxima < 50.
```

```python
print(df_empleados.loc[(df_empleados['edad']<50)])
```

```python
# -----------------------------------------------------------------------------
# EJERCICIO 6: Combinación de .loc con condiciones de grupo
# -----------------------------------------------------------------------------
# Mismo dataset df_empleados

# Tarea 6.1: Calcula el salario promedio por departamento y luego, usando .loc,
# selecciona los empleados cuyo salario sea superior al promedio de su propio departamento.
# Resultado esperado: Filas de empleados con salario > media departamental.

# Pista: puedes usar groupby + transform('mean') para crear una columna auxiliar.

# Tarea 6.2: Similar al anterior, pero selecciona únicamente los empleados del departamento 'IT'
# cuya experiencia supere la mediana de experiencia de todo el departamento 'IT'.
# Resultado esperado: Empleados de IT con experiencia > mediana de IT.
```

```python
# Tarea 6.1: Calcula el salario promedio por departamento y luego, usando .loc,
# selecciona los empleados cuyo salario sea superior al promedio de su propio departamento.
# Resultado esperado: Filas de empleados con salario > media departamental.
```
