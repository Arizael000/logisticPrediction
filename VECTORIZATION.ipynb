Plan de Aprendizaje: De "Bucles" a "Vectorización"
Fase 1: El concepto (NumPy)
El objetivo aquí es entender que las operaciones matemáticas en vectores son "element-wise" (elemento a elemento).

Concepto clave: En lugar de iterar, aplica la operación al objeto completo.

Por qué importa: La velocidad. Una operación vectorizada puede ser de 100 a 1000 veces más rápida que un bucle for en Python puro.

Fase 2: Aplicación en Datos (Pandas)
Pandas está construido sobre NumPy. Aprende a aplicar funciones a columnas enteras en lugar de usar .apply() (que a veces es solo un bucle oculto) o iterar sobre filas.

Concepto clave: Operaciones aritméticas, filtrado booleano y métodos vectorizados de strings.

Fase 3: Práctica de "Condicionales"
Aprender a reemplazar estructuras if-else dentro de bucles por funciones como np.where() o np.select().

1. Comparativa de velocidad (NumPy)
Python
import numpy as np
import time

# Crear un array grande
data = np.arange(1_000_000)

# Método lento (Bucle)
start = time.time()
resultado_bucle = []
for i in data:
    resultado_bucle.append(i * 2)
print(f"Tiempo con bucle: {time.time() - start:.4f}s")

# Método rápido (Vectorizado)
start = time.time()
resultado_vec = data * 2
print(f"Tiempo vectorizado: {time.time() - start:.4f}s")
2. Sustitución de if-else con np.where
Python
import pandas as pd

# Supongamos que queremos categorizar valores
df = pd.DataFrame({'valores': np.random.randint(0, 100, 10)})

# Manera NO vectorizada (iterando)
def categorizar(val):
    return 'Alto' if val > 50 else 'Bajo'

# Manera Vectorizada (La forma correcta)
df['categoria'] = np.where(df['valores'] > 50, 'Alto', 'Bajo')
print(df)
