```python
"""
Exercise 6: Set datetime as index and sort
- Create a DataFrame with dates: '2023-05-15', '2023-03-10', '2023-08-22' and values [10,20,30]
- Set the 'date' column as the index (after converting to datetime)
- Sort the index and display the DataFrame
"""
```

```python
df = pd.DataFrame({
    'date' : ['2023-05-15', '2023-03-10', '2023-08-22'],
    'numbers' : [10,20,30]
})
df['date'] = pd.to_datetime(df['date'])
# 1. Establecemos el índice Y 2. Ordenamos el índice
df = df.set_index('date').sort_index()
print(df)
```

```python
"""
Exercise 7: Resample time series data
- Create a DataFrame with a daily date range from '2023-01-01' to '2023-12-31' and a cumulative sum of random numbers
- Resample the data to monthly frequency, computing the mean for each month
- Also compute the sum per quarter
"""
```

```python
days =  pd.date_range(start =  '2023-01-01', end= '2023-12-31')

df = pd.DataFrame({
    'date' : days,
    'numbers': np.random.randint(1, 10, size=len(days)) # Números pequeños para ver la suma clara
})

# 2. Creamos la columna de suma acumulativa
df['acumulado'] = df['numbers'].cumsum()
```
