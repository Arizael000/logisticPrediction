# Análisis del negocio

Al conversar con el cliente (la empresa) llegamos a la conclusión de que la pregunta y objetivo está en la siguiente frase: 
"Implementar un modelo de forecasting multivariado para predecir la demanda mensual a nivel de SKU (producto individual)   
con un horizonte de 3 meses, permitiendo reducir el quiebre de stock en un y optimizar el flujo de caja mediante inversiones anticipadas basadas en tendencia y estacionalidad."

1. Esto implica que es necesaria una predicción por mes de los productos. 
Con un horizonte temporal de 3 meses, es decir, proyectar 3 meses más del último dato.

2. La granularidad del pronóstico sería a nivel de cada producto, pero usando la categoría. No es necesario llegar a nivel de SKU en este caso.

3. La pérdida más grave en términos de finanzas la daría el overstock, debido a que los productos son alimenticios y si pasan a fecha de caducidad sería cero ganancia y cero retorno de inversión.

4. Se debe ajustar la solución teniendo esto en cuenta.

En conversaciones con el cliente nos comentan que en el Mes 1 (2023), se identificó un desbalance significativo entre la oferta y la demanda:   
el inventario inicial (2,002 unidades) superó las ventas reales (756 unidades) en un 165%.   
Este excedente de 1,246 unidades representa capital inmovilizado y costos de almacenamiento innecesarios,   
validando la necesidad de un modelo de forecasting que ajuste las compras a la demanda real.

El período de estudio comprende de '2023-01-03'  a   '2024-12-11'.

Para medir el éxito del proyecto se considerará la efectividad del modelo en el forecasting de ventas en 3 meses, 
de forma que disminuya el overstock a futuro.
