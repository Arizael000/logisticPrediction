```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Leer el archivo de Excel
# Nota: Online Retail II tiene dos pestañas (Year 2009-2010 y Year 2010-2011)
print("Leyendo el archivo Excel... (esto puede tardar un poco)")
df_1 = pd.read_excel('online_retail_II.xlsx', sheet_name='Year 2009-2010')
df_2 = pd.read_excel('online_retail_II.xlsx', sheet_name='Year 2010-2011')

# Combinar ambas pestañas en un solo DataFrame
df_total = pd.concat([df_1, df_2], ignore_index=True)

# 2. Configurar la conexión a PostgreSQL
# Reemplaza con tus credenciales: usuario, contraseña, host y nombre de tu BD
USER = 'postgres'
PASSWORD = '2305'
HOST = 'localhost'
PORT = '5432'
DB_NAME = 'datascience_db'

engine = create_engine(f'postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB_NAME}')

# 3. Cargar los datos en PostgreSQL
print("Cargando datos en PostgreSQL...")
df_total.to_sql('online_retail', con=engine, if_exists='replace', index=False)
print("¡Carga completada con éxito!")
```
