Conversión de columnas y tipos
Comprobar de que los tipos estén correctos (especialmente fecha como datetime):

```python
import pandas as pd

df = pd.read_pickle("datos_procesados.pkl")
df["fecha"] = pd.to_datetime(df["fecha"])

print((df["fecha"].dtypes))
```

Crear columnas temporales útiles
Ya que tengo datos desde marzo 2023 hasta diciembre 2024 (más de un año y medio), usaré mes, trimestre, año_mes como variables predictoras:

```python
df["año_mes"] = df["fecha"].dt.to_period("M")
df["mes"] = df["fecha"].dt.month
df["trimestre"] = df["fecha"].dt.quarter
df["año"] = df["fecha"].dt.year
print(df.head())
```

Agrupar ventas por producto y mes
Esto crea un dataset más limpio para modelar la cantidad total vendida por producto por mes:

```python
ventas_mensuales = (
    df.groupby(["producto", "año_mes", "categoria", "medida"])
    .agg({
        "cantidad": "sum",
        "total": "sum"
    })
    .reset_index()
)

# Convertir año_mes a fecha real
ventas_mensuales["fecha"] = ventas_mensuales["año_mes"].dt.to_timestamp()
ventas_mensuales = ventas_mensuales.sort_values(by=["producto", "fecha"])

print(ventas_mensuales.head())
```

Crear variables predictoras adicionales
a. Variable secuencial mes_n por productos

```python
ventas_mensuales["mes_n"] = ventas_mensuales.groupby("producto").cumcount()
print(ventas_mensuales.head())
```

b. Media móvil (opcional) es la media de los 3 meses anteriores:

```python
ventas_mensuales["media_3m"] = (
    ventas_mensuales
    .groupby("producto")["cantidad"]
    .transform(lambda x: x.rolling(window=3, min_periods=1).mean())
)
print(ventas_mensuales)
```

**Seleccionar variable Target y predictora:**

Codificar variables categóricas

```python
ventas_mensuales["producto_cod"] = ventas_mensuales["producto"].astype("category").cat.codes
ventas_mensuales["categoria_cod"] = ventas_mensuales["categoria"].astype("category").cat.codes
ventas_mensuales["medida_cod"] = ventas_mensuales["medida"].astype("category").cat.codes
print(ventas_mensuales.head())
```

```python
ventas_mensuales.to_csv("ventas_mensuales_limpio.csv", index=False)
```
