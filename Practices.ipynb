**Pruebas para aprender el sistema de la API de pandas para manipular datos en DataScience**

**Función	Descripción	Ejemplo de uso:**
sum	Suma	                            .groupby(...).sum()
mean	Promedio	                    .groupby(...).mean()
median	Mediana                         .groupby(...).median()
min	Valor mínimo	                    .groupby(...).min()
max	Valor máximo	                    .groupby(...).max()
count	Cuenta no nulos	                .groupby(...).count()
size	Cuenta total (incluye nulos)	.groupby(...).size()
std	Desviación estándar	                .groupby(...).std()
var	Varianza	                        .groupby(...).var()
first	Primer valor del grupo	        .groupby(...).first()
last	Último valor del grupo	        .groupby(...).last()
nunique	Número de valores únicos	    .groupby(...).nunique()
prod	Producto de los valores	        .groupby(...).prod()
sem	Error estándar de la media	        .groupby(...).sem()
mad	Desviación absoluta de la mediana	.groupby(...).mad() (menos común)

Ejemplo1:

```python
import pandas as pd

data = {
    "producto": ["A", "A", "B", "B", "C", "C"],
    "categoria": ["x", "x", "y", "y", "z", "z"],
    "cantidad": [10, 20, 15, 5, 8, 12],
    "precio_unitario": [100, 100, 200, 200, 150, 150]
}

ventas = pd.DataFrame(data)
ventas["total"] = ventas["cantidad"] * ventas["precio_unitario"]
print(ventas)
```

 **1. Suma total por producto**

```python
print(ventas.groupby("producto")["total"].sum()
)
```

 **2. Media de cantidad por categoría**

```python
print(ventas.groupby("categoria")["cantidad"].mean())
```

**3. Agregaciones múltiples**

```python
df1=ventas.groupby("producto").agg({
    "cantidad": "sum",
    "total": ["mean", "sum"]
})

# Aplanar columnas OJOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOO APLANAS COLUMNAS CUANDO HAGO múltiples agregaciones
df1.columns = ['_'.join(col) for col in df1.columns]
df1 = df1.reset_index()
print(df1)
```

**4. Agrupaciones por múltiples columnas**

```python
print(ventas.groupby(["categoria", "producto"])["total"].median())
```
