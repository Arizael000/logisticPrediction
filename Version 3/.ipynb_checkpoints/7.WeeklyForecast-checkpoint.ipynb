# 7. WeeklyForecast - Forecasting Semanal con Prophet (v2)

## Motivacion
ARIMA diario tenia SMAPE ~50%. Prophet semanal reduce ruido y mejora precision.

## Cambios v2
- Bugfix expansion diaria: usar datos diarios reales para perfil de dia de semana
- Prophet con changepoint_prior_scale=0.01 (mas conservador, menos sobreajuste)
- Seasonalities: yearly=False (solo 2 anos de datos, riesgo sobreajuste)


```python
import pandas as pd, numpy as np, warnings, unicodedata
from datetime import datetime, timedelta
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from prophet import Prophet
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
np.random.seed(42)
print("Imports OK")
```

```python
df = pd.read_csv('clean_data.csv')
if 'Unnamed: 0' in df.columns: df = df.drop(columns='Unnamed: 0')
df['fecha'] = pd.to_datetime(df['fecha'])
df = df.set_index('fecha').sort_index()
print(f"Data: {len(df)} rows, {df.index.min().date()} a {df.index.max().date()}")
categorias = sorted(df['categoria'].unique().tolist())
print(f"Categorias ({len(categorias)}): {categorias}")
```

```python
# Agregacion semanal (W-MON: semanas terminando en lunes)
dfs_weekly = {}
for cat in categorias:
    s = df[df['categoria'] == cat]['cantidad'].resample('W-MON').sum()
    s.name = 'cantidad'
    dfs_weekly[cat] = s.to_frame()

for cat, d in dfs_weekly.items():
    print(f"{cat:20s} | {len(d):3d} semanas | {d.index.min().date()} a {d.index.max().date()} | "
          f"Prom={d['cantidad'].mean():.0f}/sem | Total={d['cantidad'].sum():.0f}")

N_TEST_W = 13  # 13 semanas = ~91 dias
```

```python
# Train/Test split semanal
train_w = {}; test_w = {}
for cat in categorias:
    d = dfs_weekly[cat].sort_index()
    split = len(d) - N_TEST_W
    train_w[cat] = d.iloc[:split].copy()
    test_w[cat] = d.iloc[split:].copy()
print(f"Train/Test split: {len(train_w[categorias[0]])} train, {N_TEST_W} test semanas")
```

```python
# Funciones de evaluacion
def smape(y_true, y_pred):
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    mask = denom != 0
    if mask.sum() == 0: return 0.0
    return np.mean(np.abs(y_true[mask] - y_pred[mask]) / denom[mask]) * 100

def expandir_a_diario(weekly_forecast, weekly_ci_l, weekly_ci_u, cat, df_daily, n_weeks=13):
    """Expande forecast semanal a diario usando perfil historico REAL por dia de semana.
    df_daily: DataFrame diario original con columna 'cantidad' indexado por fecha"""
    cat_daily = df_daily[df_daily['categoria'] == cat]['cantidad']
    dow_profile = cat_daily.groupby(cat_daily.index.dayofweek).mean()
    dow_profile = dow_profile / dow_profile.sum()

    daily_fc, daily_ci_l, daily_ci_u = [], [], []
    for w_idx in range(n_weeks):
        week_total = weekly_forecast[w_idx]
        ci_l_w = weekly_ci_l[w_idx]
        ci_u_w = weekly_ci_u[w_idx]
        for dow in range(7):
            prop = dow_profile.get(dow, 1/7)
            daily_fc.append(week_total * prop)
            daily_ci_l.append(ci_l_w * prop)
            daily_ci_u.append(ci_u_w * prop)

    return np.array(daily_fc), np.array(daily_ci_l), np.array(daily_ci_u)

print("Funciones OK")
```

```python
# ============================================
# ENTRENAMIENTO: Prophet
# ============================================
# Nota: ETS descartado - requiere >2 ciclos estacionales completos con datos semanales.
# Usamos Prophet con parametros conservadores (changepoint_prior_scale=0.01)
# y yearly_seasonality=False porque con solo 2 anos se sobreajusta.

print("=== Prophet (conservador) ===")
modelos_prophet = {}
for cat in categorias:
    train_df = train_w[cat].reset_index()[['fecha', 'cantidad']]
    train_df.columns = ['ds', 'y']

    try:
        model = Prophet(
            yearly_seasonality=False,
            weekly_seasonality=False,
            daily_seasonality=False,
            seasonality_mode='additive',
            changepoint_prior_scale=0.01,
            changepoint_range=0.8
        )
        model.fit(train_df)

        future = model.make_future_dataframe(periods=N_TEST_W, freq='W-MON')
        forecast = model.predict(future)
        pred = np.maximum(forecast['yhat'].values[-N_TEST_W:], 0)
        ci_l = np.maximum(forecast['yhat_lower'].values[-N_TEST_W:], 0)
        ci_u = np.maximum(forecast['yhat_upper'].values[-N_TEST_W:], 0)

        true = test_w[cat]['cantidad'].values
        s = smape(true, pred)
        print(f"  {cat:20s} | SMAPE: {s:.1f}%")
        modelos_prophet[cat] = (model, pred, ci_l, ci_u, s)
    except Exception as e:
        print(f"  {cat:20s} | ERROR: {e}")
        modelos_prophet[cat] = None
```

```python
# ============================================
# RESULTADOS EN TEST
# ============================================
print("\n" + "="*80)
print("RESULTADOS EN TEST (13 semanas)")
print("="*80)

resultados = []
for cat in categorias:
    if modelos_prophet[cat] is None: continue
    model, pred, ci_l, ci_u, s = modelos_prophet[cat]
    true = test_w[cat]['cantidad'].values

    resultados.append({'Categoria': cat, 'SMAPE': round(s, 1),
                       'Test_Total': round(true.sum(), 0),
                       'Pred_Total': round(pred.sum(), 0)})

    print(f"  {cat:20s} | SMAPE: {s:.1f}% | Test={true.sum():.0f} | Pred={pred.sum():.0f}")

print("\n=== TABLA COMPARATIVA ===")
print(pd.DataFrame(resultados).to_string(index=False))
```

```python
# ============================================
# FORE CAST 13 SEMANAS (91 DIAS)
# ============================================
print("\n" + "="*80)
print("FORECAST SEMANAL 13 SEMANAS")
print("="*80)

forecast_weekly = {}
for cat in categorias:
    model = modelos_prophet[cat][0]
    future = model.make_future_dataframe(periods=13, freq='W-MON')
    fcst = model.predict(future)
    fc = np.maximum(fcst['yhat'].values[-13:], 0)
    ci_l = np.maximum(fcst['yhat_lower'].values[-13:], 0)
    ci_u = np.maximum(fcst['yhat_upper'].values[-13:], 0)

    forecast_weekly[cat] = {'fc': fc, 'ci_l': ci_l, 'ci_u': ci_u}
    print(f"  {cat:20s} | Sem1={fc[0]:.0f} Sem13={fc[-1]:.0f} Total={fc.sum():.0f} | "
          f"Prom/hist={(dfs_weekly[cat]['cantidad'].mean()):.0f}/sem")
```

```python
# ============================================
# EXPANSION A DIARIO Y EXPORTACION CSV
# ============================================
# BUGFIX v2: pasar df (datos diarios) en vez de dfs_weekly a expandir_a_diario
print("\n" + "="*80)
print("EXPANSION A DIARIO Y EXPORTACION CSV")
print("="*80)

last_date = df.index.max()
fecha_inicio = last_date + timedelta(days=1)

for cat in categorias:
    wfc = forecast_weekly[cat]
    daily_fc, daily_ci_l, daily_ci_u = expandir_a_diario(
        wfc['fc'], wfc['ci_l'], wfc['ci_u'], cat, df, n_weeks=13)

    fechas = [fecha_inicio + timedelta(days=i) for i in range(91)]

    df_out = pd.DataFrame({
        'fecha': fechas,
        'cantidad_forecast': daily_fc,
        'ci_lower': daily_ci_l,
        'ci_upper': daily_ci_u,
        'categoria': cat
    })

    nfkd = unicodedata.normalize('NFKD', cat)
    clean_cat = nfkd.encode('ascii', 'ignore').decode('ascii').replace(' ', '_')
    filename = f'forecast_{clean_cat}_90d.csv'
    df_out.to_csv(filename, index=False)
    print(f"  {cat:20s} -> {filename} | Total 91d: {daily_fc.sum():.0f} (vs semanal {wfc['fc'].sum():.0f})")

# V2: verificar que expansion no infla
print("\n[VERIFICACION] Suma semanal vs suma expandida:")
for cat in categorias:
    wfc = forecast_weekly[cat]
    daily_fc, _, _ = expandir_a_diario(wfc['fc'], wfc['ci_l'], wfc['ci_u'], cat, df, n_weeks=13)
    ratio = round(daily_fc.sum() / wfc['fc'].sum(), 2) if wfc['fc'].sum() > 0 else 0
    ok = "OK" if 0.99 <= ratio <= 1.01 else "BUG!"
    print(f"  {cat:20s} | semanal={wfc['fc'].sum():.0f} | expandido={daily_fc.sum():.0f} | ratio={ratio} {ok}")

# Consolidado
all_fc = []
for cat in categorias:
    wfc = forecast_weekly[cat]
    daily_fc, daily_ci_l, daily_ci_u = expandir_a_diario(
        wfc['fc'], wfc['ci_l'], wfc['ci_u'], cat, df, n_weeks=13)
    for i in range(91):
        all_fc.append({
            'fecha': fecha_inicio + timedelta(days=i),
            'categoria': cat,
            'cantidad_forecast': daily_fc[i],
            'ci_lower': daily_ci_l[i],
            'ci_upper': daily_ci_u[i]
        })

df_con = pd.DataFrame(all_fc)
df_con.to_csv('forecast_consolidado_90d.csv', index=False)
total = df_con['cantidad_forecast'].sum()
print(f"\nCONSOLIDADO: forecast_consolidado_90d.csv ({total:.0f} uds)")
```

```python
# ============================================
# COMPARACION CON MODELO ORIGINAL (ARIMA DIARIO)
# ============================================
print("\n" + "="*80)
print("COMPARACION: ARIMA Diario vs Prophet Semanal (v2)")
print("="*80)

arima_results = {
    'Bebidas': 48.7, 'Carnes': 41.3, 'Congelados': 57.5,
    'Descartables': 43.9, 'Embutidos': 68.6, 'Enlatados': 54.1,
    'Lácteos': 42.5
}

print(f"{'Categoria':20s} {'ARIMA(diario)':15s} {'Prophet(v2)':15s} {'Mejora':10s}")
print("-"*60)
n_mejora = 0
n_empeora = 0
for r in resultados:
    cat = r['Categoria']
    arima_smape = arima_results.get(cat, 0)
    new_smape = r['SMAPE']
    mejora = arima_smape - new_smape
    if mejora > 1: n_mejora += 1
    elif mejora < -1: n_empeora += 1
    tag = "MEJORO" if mejora > 2 else ("EMPEORO" if mejora < -2 else "~")
    print(f"  {cat:20s} | {arima_smape:5.1f}%{'':8s} | {new_smape:5.1f}%{'':8s} | {mejora:+5.1f}% ({tag})")

print("-"*60)
avg_arima = np.mean(list(arima_results.values()))
avg_new = np.mean([r['SMAPE'] for r in resultados])
print(f"  {'PROMEDIO':20s} | {avg_arima:.1f}%{'':8s} | {avg_new:.1f}%{'':8s} | {avg_arima-avg_new:+5.1f}%")
print(f"  Mejoraron: {n_mejora}/{len(resultados)} | Empeoraron: {n_empeora}/{len(resultados)}")
```

```python
# ============================================
# GRAFICO COMPARATIVO
# ============================================
fig, axes = plt.subplots(4, 2, figsize=(16, 14))
axes = axes.flatten()
for idx, cat in enumerate(categorias):
    ax = axes[idx]
    hist_w = dfs_weekly[cat]['cantidad']
    ax.plot(range(len(hist_w)), hist_w.values, 'o-', label='Semanal Historico', alpha=0.6, ms=4, color='blue')
    wfc = forecast_weekly[cat]
    daily_fc, daily_ci_l, daily_ci_u = expandir_a_diario(
        wfc['fc'], wfc['ci_l'], wfc['ci_u'], cat, df)
    x_daily = range(len(hist_w), len(hist_w) + 91)
    ax.plot(x_daily, daily_fc, 's--', label='Forecast Semanal (expandido)', alpha=0.8, ms=3, color='green', lw=2)
    ax.fill_between(x_daily, daily_ci_l, daily_ci_u, alpha=0.15, color='green', label='IC 95%')
    ax.axvline(x=len(hist_w) - N_TEST_W, color='red', ls=':', alpha=0.5, label='Inicio test')

    # Mostrar promedios historico y forecast
    hist_mean = hist_w.mean()
    fc_mean = wfc['fc'].mean()
    ax.axhline(y=hist_mean, color='gray', ls='--', alpha=0.4, label=f'Prom Hist ({hist_mean:.0f}/sem)')
    ax.set_title(f'{cat} | SMAPE Test: {resultados[idx]["SMAPE"]:.1f}%', fontsize=11)
    ax.legend(fontsize=7)
    ax.grid(True, alpha=0.3)
    ax.set_ylabel('Unidades')
axes[-1].axis('off')
plt.suptitle('Forecast Semanal con Prophet v2 (13 semanas)', fontsize=14)
plt.tight_layout()
plt.savefig('comparacion_semanal.png', dpi=150, bbox_inches='tight')
plt.show()
print("Grafico guardado: comparacion_semanal.png")
```
