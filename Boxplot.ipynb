Esta notebook es solo para aprendizaje y práctica de Boxplots

```python
#Código tomado de Gemini (Razonamiento, free version) for educational purposes
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Simulamos unos datos parecidos a los tuyos
data = {
    'categoria': ['Panadería']*50 + ['Granos']*50 + ['Lácteos']*50,
    'cantidad': np.concatenate([
        np.random.normal(180, 20, 50), # Panadería: media 180
        np.random.normal(150, 10, 50), # Granos: media 150 (más estable)
        np.random.normal(100, 50, 50)  # Lácteos: media 100 (mucha dispersión)
    ])
}
df = pd.DataFrame(data)

# Crear el Boxplot
plt.figure(figsize=(10, 6))
# Asignamos 'categoria' a 'hue' para los colores y quitamos la leyenda innecesaria
sns.boxplot(x='categoria', y='cantidad', data=df, hue='categoria', palette='Set2', legend=False)

plt.title('Distribución de Ventas por Categoría', fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Código para comparar Normal vs. No Normal

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Creamos datos: uno Normal y uno Sesgado (Ventas)
np.random.seed(42)
datos_normales = np.random.normal(loc=100, scale=20, size=1000)
datos_sesgados = np.random.exponential(scale=50, size=1000) # Típico de ventas

df_comp = pd.DataFrame({
    'Distribución': ['Normal'] * 1000 + ['Sesgada (Ventas)'] * 1000,
    'Valores': np.concatenate([datos_normales, datos_sesgados])
})

# Graficamos
plt.figure(figsize=(10, 6))
sns.boxplot(x='Distribución', y='Valores', data=df_comp, hue='Distribución', palette='Set1', legend=False)

plt.title('¿Cómo se ve la No Normalidad?', fontsize=15)
plt.show()
```
